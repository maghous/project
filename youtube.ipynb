```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
```

```python
data=pd.read_csv("top-300-youtube-channels.csv")
data.info()
```

```python
data.tail()
```

```python
data.drop(columns=["Unnamed: 0","Rank"],inplace=True)
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
text=data["Genre"].unique()
wordcloud = WordCloud(max_words=10000,background_color="red").generate(str(text))
plt.rcParams['figure.figsize'] = (12, 20)
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis("off")
plt.show()
```

```python
text=data["Channel_Name"].unique()
wordcloud = WordCloud(max_words=10000,background_color="red").generate(str(text))
plt.rcParams['figure.figsize'] = (12, 20)
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis("off")
plt.show()
```

```python
plt.scatter(data['Channel_Name'],data["Channel_Started"],color="red");
```

```python
sns.set(style="whitegrid", font_scale=1.2, rc={"figure.figsize":(20,8)})
sns.countplot(data=data,x='Channel_Started',color='red',saturation=0.35)
plt.xlabel('Channel Count',fontsize=15)
plt.ylabel('Years',fontsize=15)
plt.title("Number of Channels created per Year",fontsize=20)
```
