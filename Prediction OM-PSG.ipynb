Let is assumed that the number of goals scored by a team during a match follows a law of Poisson of parameter Lambda_team specific to the team.

we create a function that takes a team name as input and returns an estimate of the parameter lambda_team.

```python
def LambdaEquipe(Equipe):
    datHome = df[df.HomeTeam==Equipe].FTHG
    datAway = df[df.AwayTeam==Equipe].FTAG
    return (np.mean(datHome) + np.mean(datAway))/2
LambdaEquipe("St Etienne")
```

Then we create a function which takes as input two parameters $\mu$,$\nu$ and returns the vecotor of probabilities $(\mathbb{P}(X>Y),\mathbb{P}(X=Y),\mathbb{P}(X<Y))$ $\newline$
Knowing that $X\sim \mathcal{P}(\mu)$ and $Y\sim \mathcal{P}(\nu)$

```python
def CompPoisson(mu, nu):
    m = np.arange(np.ceil(10*mu))
    n = np.arange(np.ceil(10*nu))
    prob1 = sp.poisson.pmf(m, mu)
    prob2 = sp.poisson.pmf(n, nu)
    probproduct = prob1 * prob2[:, None]
    return np.sum(np.triu(probproduct, 1)), np.sum(np.diagonal(probproduct)), np.sum(np.tril(probproduct, -1))

CompPoisson(1,2)
```

finally we create a function that takes two team names as input and returns an estimate of the probability of victory, draw and defeat of a match between the two teams.

```python
def PredictionFoot(Equipe1, Equipe2):
    return CompPoisson(LambdaEquipe(Equipe1), LambdaEquipe(Equipe2))
```

the first element represent the probability of victory, the second one the probability of draw and the last the probability of defeat.

```python
p = PredictionFoot("Paris SG", "Marseille")
print("Victory PSG: {:.2f}%".format(100*p[0]))
print("Draw: {:.2f}%".format(100*p[1]))
print("Defeat OM: {:.2f}%".format(100*p[2]))
```
