```python
dt=dt[dt["Appearances"]!=0]
```

```python
dt.shape
```

```python
goalscorers=dt.nlargest(5,["Goals"])
dims=(11.7,8.27)
fig,ax=plt.subplots(figsize=dims)
sns.barplot(ax=ax,x="Name",y="Goals",data=goalscorers,).set(title="Meilleur buteur")
plt.xlabel("Player");
```

```python
playmakers=dt.nlargest(5,['Assists'])
```

```python
playmakers
```

```python
fig,ax=plt.subplots(figsize=dims)
sns.barplot(ax=ax,x="Name",y="Assists",data=playmakers).set(title="Meilleurs passeur")
plt.xlabel("Player");
```

```python
dt["Involvements"]=dt["Goals"] + dt["Assists"]
involvements=dt.nlargest(5,'Involvements')
```

```python
fig,ax=plt.subplots(figsize=dims)
sns.barplot(ax=ax,x="Name",y="Involvements",data=involvements).set(title="les joueurs plus participé dans les occasions")
plt.xlabel("Player")
plt.ylabel("Goal Involvements");
```
