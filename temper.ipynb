```python
plt.figure(figsize=(20,12))
plt.title("cor-relation")
sns.heatmap(data.corr(),annot=True,cmap='viridis')
plt.show()
```

```python
def reg(data,col,title='',ax=None):

    sns.regplot(data=data,x='YEAR',y=col,scatter=True,fit_reg=True,logistic=False,ax=ax)
    ax.set_title(f"regresion:{title}")
```

```python
fig,axes=plt.subplots(4,3,figsize=(30,30))
reg(data,'JAN',title='jan',ax=axes[0,0])


reg(data,'FEB',title='feb',ax=axes[0,1])
reg(data,'MAR',title='mar',ax=axes[0,2])
reg(data,'APR',title='apr',ax=axes[1,0])
reg(data,'MAY',title='may',ax=axes[1,1])

reg(data,'JUN',title='jun',ax=axes[1,2])
reg(data,'JUL',title='jul',ax=axes[2,0])
reg(data,'AUG',title='aug',ax=axes[2,1])
reg(data,'SEP',title='sep',ax=axes[2,2])

reg(data,'OCT',title='oct',ax=axes[3,0])
reg(data,'NOV',title='nov',ax=axes[3,1])
reg(data,'DEC',title='dec',ax=axes[3,2])
```

```python
def plot(data,col,ax=None):
    sns.jointplot(data=data,x='YEAR',y=col,ax=ax)
```

```python
plot(data,'JAN-FEB')
plt.show()
```

```python
plot(data,'MAR-MAY')
```
