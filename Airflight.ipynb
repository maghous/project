```python
sns.kdeplot(data=data,x="Age",hue="Flight Status");
```

```python
data["Nationality"].value_counts().head(10).plot.bar(color="red")
plt.title("Top 10 passenger nationalities");
```

```python
plt.figure(figsize=(10,10))
data["Nationality"].value_counts().head(20).plot.pie();
```

```python
data["Continents"].value_counts().plot.bar(color="red")
plt.title("the proportion of passengers for each continent");
```

```python
plt.figure(figsize=(10,10))
data["Continents"].value_counts().plot.pie();
```

```python
data["Flight Status"].value_counts().head(10).plot.bar(color="red")
plt.title("the proportion of Flight status");
```

```python
plt.figure(figsize=(10,10))
data["Flight Status"].value_counts().head(10).plot.pie();
```

```python
data["Country Name"].value_counts().head(10).plot.bar(color="red")
plt.title('The proportion of top 10 flight departures countries');
```
